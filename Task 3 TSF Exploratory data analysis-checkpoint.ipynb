# Profit vs Discount

```python
fig,ax=plt.subplots(figsize=(25,8))
ax.scatter(data['Sales'],data['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=data)
plt.legend()
plt.show()
print("Observation 4: No correlation between profit and discoun")
```

# Profit vs Quantity

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=data)
plt.legend()
plt.show()
```

```python
data.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['red','orange'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
print("Observation 5: Profit and sales are maximum in consumer segment and minimum in Home Office segment")
```

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=data,hue='Segment',order=data['Region'].value_counts().index,palette='colorblind')
plt.xlabel('Region',fontsize=15)
plt.show()
print("Observation 6: Segment wise sales are almost same in every region")
```

```python
data.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['red','orange'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
print("Observation 7: Profit and sales are maximum in west region and minimum in south region")
```

```python
ps = data.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['orange','red'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
print("Observation 8:")
print(".high profit is for california, new york \n.loss is for texas, pennsylvania, Ohio")
```

```python
r_states = data['State'].value_counts().nlargest(10)
r_states
```

```python
data.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['red','orange'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
print("Observation 9:")
print("As a business manager, try to find out the weak areas where you can work to make more profit?")
print(".Technology and Office Supplies have high profit\n.Furniture have less profit")
```

```python
ps = data.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','orange'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
print("Observation 10:")
print(".Phones sub-category have high sales\n.Chairs have high sales but less profit compared to phones\n.Tables and Bookmarks sub-categories facing huge loss")
```
